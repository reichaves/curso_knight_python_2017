```python
from selenium import webdriver
from selenium.common.exceptions import NoAlertPresentException
from selenium.webdriver.support.select import Select
import pandas as pd
from bs4 import BeautifulSoup
```

Exemplo de arquivo com nomes de políticcos - deputados eleitos em 2014

```python
eleitos_d = pd.read_excel('deputados_eleitos_2014.xlsx',sheet_name='Sheet1', converters={'politico_cpf': lambda x: str(x)})
```

```python
eleitos_d.info()
```

```python
# Lista para receber o dataframe final
links = []
```

```python
def busca_rede_social(busca):
    profile = webdriver.FirefoxProfile()
    browser = webdriver.Firefox(profile)
    browser.implicitly_wait(10)
    
    browser.get(busca)
    
    html = browser.page_source
    soup = BeautifulSoup(html, "html.parser")
    browser.close()
    
    for link in soup.find_all("a"):
        try: 
            href = link['href']
            if 'twitter.com' in href:
                if 'status' in href:
                    return 'Not Found'
                else:
                    return href
        except KeyError:
            continue
```

```python
for num, row in eleitos_d.iterrows():
    cpf = (row['politico_cpf']).strip()
    deputado = (row['politico_nome_urna']).strip()
    print(deputado)
    cargo = (row['cargo']).strip()
    
    nome = deputado.replace(' ', '+')
    
    # Cria url para busca - exemplo do Twitter
    busca = f'https://www.google.com.br/search?q=twitter+{nome}+{cargo}'

    rede_social = busca_rede_social(busca)
    
    links.append({'cpf': cpf, 
                  'deputado': deputado, 
                  'linK_twitter': rede_social})
```

```python
df_twitter = pd.DataFrame(links, columns = ['cpf', 'deputado', 'linK_twitter'])
```
