```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
jornais_no_twitter = ['folha', 'estadao', 'jornaloglobo', 'zerohora']
```

```python
def seguidores(usuario_twitter):
	consulta_api = api.get_user(usuario_twitter)._json
	seguidores = consulta_api['followers_count']
	return seguidores
```

```python
seguidores('paidatocandeira')
```

```python
for jornal in jornais_no_twitter:
    numero = seguidores(jornal)
    print(f'O usu√°rio {jornal} tem {numero} seguidores no Twitter')
```
