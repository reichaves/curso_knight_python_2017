```python
import tweepy
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import json
```

```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
# Procurei formas de fazer aqui (https://stackoverflow.com/questions/25588272/return-a-users-tweets-with-tweepy) e aqui (http://docs.tweepy.org/en/v3.5.0/api.html)
```

```python
my_ids = ['jeanwyllys_real', 'lucianagenro', 'jairbolsonaro', 'MBLivre']
dict_of_tweets = []
cant_find_tweets_for_those_ids = []

for each_id in my_ids:   
    try:
        tweets = api.user_timeline(each_id)
        for each_json_tweet in tweets:
            dict_of_tweets.append(each_json_tweet._json)
    except Exception as e:
        cant_find_tweets_for_those_ids.append(each_id)

with open('tweet_json_bolhas.txt', 'w') as file:
        file.write(json.dumps(dict_of_tweets, indent=4))
```

```python
my_demo_list = []
with open('tweet_json_bolhas.txt', encoding='utf-8') as json_file:  
    all_data = json.load(json_file)
    for each_dictionary in all_data:
        nome = each_dictionary['user']['name']
        tweet_id = each_dictionary['id']
        text = each_dictionary['text']
        favorite_count = each_dictionary['favorite_count']
        retweet_count = each_dictionary['retweet_count']
        created_at = each_dictionary['created_at']
        my_demo_list.append({'nome': str(nome),
                             'tweet_id': str(tweet_id),
                             'text': str(text),
                             'favorite_count': int(favorite_count),
                             'retweet_count': int(retweet_count),
                             'created_at': created_at,
                            })
        #print(my_demo_list)
        tweet_json = pd.DataFrame(my_demo_list, columns = 
                                  ['nome', 'tweet_id', 'text', 
                                   'favorite_count', 'retweet_count', 
                                   'created_at'])
```

```python
tweet_json.reset_index()
```
