```python
def buscar_deputados():
    deputados = []
    for n in range(1,7):
        url = f'https://dadosabertos.camara.leg.br/api/v2/deputados?itens=100&pagina={n}'
        r = requests.get(url)
        for linha in r.json()['dados']:
            deputados.append(linha)
    return deputados
```

```python
deputados = buscar_deputados()
```

```python
def encontrar_no_twitter(nome, cargo):
    nome = nome.replace(' ', '+')
    #### PRECISA TIRAR APOSTROFO OU OUTROS SIMBOLOS DO NOME
    busca = f'https://www.google.com.br/search?q=twitter+{nome}+{cargo}'
    #print(busca)
    browser = webdriver.Chrome(chrome_path)
    browser.get(busca)
    html = browser.page_source
    soup = BeautifulSoup(html, "html.parser")
    browser.close()
    
    
    for link in soup.find_all("a"):
        try: 
            href = link['href']
            if 'twitter.com' in href:
                if 'status' in href:
                    return 'Not Found'
                else:
                    return href
        except KeyError:
            continue
```
